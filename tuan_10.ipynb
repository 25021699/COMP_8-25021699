W8A3: ValueError - HappyNumbers
Nhập vào một số nguyên dương a, Kiểm tra xem a có phải là số hạnh phúc không. Một số nguyên dương n được gọi là số hạnh phúc nếu ta thực hiện liên tiếp các bước sau:
- Tính tổng bình phương các chữ số của n.
- Lặp lại bước đó với kết quả vừa tính.
- Nếu cuối cùng kết quả là 1, thì n là số hạnh phúc.
- Nếu rơi vào vòng lặp vô hạn không bao giờ ra 1 (ví dụ như 4 → 16 → 37 → ...), thì n không phải là số hạnh phúc.
Hãy nhập thử đầu vào là "a" rồi đọc kỹ Traceback để gỡ lỗi nhập liệu. (Dùng try ... except bắt lỗi ValueError thêm set)


```python
def is_happy_number(n: int) -> bool:
    seen = set() 
    while n != 1 and n not in seen:
        seen.add(n)
        s = 0
        while n > 0:
            digit = n % 10
            s += digit * digit
            n //= 10
        n = s
    return n == 1
try:
    a = int(input("Nhap so nguyen duong a: "))
    if a <= 0:
        print("a khong phai so nguyen duong")
    else:
        if is_happy_number(a):
            print("a la so hanh phuc")
        else:
            print("a khong phai so hanh phuc")

except ValueError:
    print("Gia tri nhap vao khong hop le (khong phai so nguyen)")
```

W8A4:
Sử dụng try except,kết hợp cùng else... finally. Tính tổng 2 số, nếu nhập sai kiểu thì thông báo, luôn luôn in "Kết thúc chương trình." sau khi chạy

```python
try:
    a = float(input("Nhap so thu nhat: "))
    b = float(input("Nhap so thu hai: "))
except ValueError:
    print("Nhap sai kieu du lieu")
else:
    print("Tong hai so la:", a + b)
finally:
    print("Ket thuc chuong trinh.")
```

W10A10:
Viết chương trình nhập vào phần tử N (N > 0) và một mảng gồm N số nguyên dương. Một mảng là hợp lệ nếu nó gồm các số khác nhau. In ra tổng các phần tử nếu mảng đó là hợp lệ. Nếu mảng không hợp lệ, in ra "Mang khong hop le".


```python
try:
    n=int(input())
    lst=list(map(int, input().split()))
    k = list(set(lst))
    if all(i>= 0 for i in lst) and len(k) == n:
        print(sum(lst))
    else:
        print("Mang khong hop le")
except:
    print("Mang khong hop le")
```

W10A11:
Cho string S, kiểm tra xem S có phải một regex hợp lệ hay không. Đầu vào: Dòng đầu tiên chứa số nguyên T - số lượng testcase. T dòng tiếp theo mỗi dòng chứa một chuỗi S.
Output: In ra "Valid" hoặc "Invalid" cho mỗi testcase.
Ví dụ:
2
.*\+
.*+
Sample Output:
True
False
Giải thích:
.*+ : lỗi "multiple repeat" - invalid


```python
import re

T = int(input())
for _ in range(T):
    S = input().strip()
    try:
        re.compile(S)
        print("True")
    except re.error:
        print("False")
```
