```python
# W7A1
def binary_search(arr,target):
    left, right = 0, len(arr) - 1
    while left <= right:
        mid = (left + right) // 2
        if arr[mid] == target:
            return mid
        elif arr[mid] < target:
            left = mid + 1
        else:
            right = mid - 1
    return -1
arr = list(map(int,input().split()))
x = int(input())
print(binary_search(arr,x))
```

```python
# W7A2
def count_occur(arr,x):
    cnt = 0
    for i in arr:
        if i == x:
            cnt += 1
    return cnt

arr = list(map(int,input().split()))
x = int(input())
print(count_occur(arr,x))
```

```python
#W7A3
def bb_sort(arr):
    n = len(arr)
    for i in range(n - 1):
        for j in range(n - 1 - i):
            if arr[j] > arr[j + 1]:
                arr[j], arr[j + 1] = arr[j + 1], arr[j]
    return arr

def sl_sort(arr):
    n = len(arr)
    for i in range(n - 1):
        min_index = i
        for j in range(i + 1, n):
            if arr[j] < arr[min_index]:
                min_index = j
        arr[i], arr[min_index] = arr[min_index], arr[i]
    return arr

a = list(map(int,input().split()))
print(bb_sort(a))
```

```python
#W7A4
def most_frequent(arr):
    max_count = 0
    result = arr[0]

    for i in range(len(arr)):
        count = 0
        for j in range(len(arr)):
            if arr[j] == arr[i]:
                count += 1

        if count > max_count:
            max_count = count
            result = arr[i]

    return print(f"{result} xuất hiện nhiều nhất, sớm nhất,{max_count} lần")

a = list(map(int,input().split()))
most_frequent(a)
```

```python
# W7A5
def count_pairs(arr, X):
    count = 0
    n = len(arr)

    for i in range(n - 1):
        for j in range(i + 1, n):
            if arr[i] + arr[j] == X:
                count += 1

    return count

arr = list(map(int,input().split()))
x = int(input())
print(count_pairs(arr,x))
```

```python
# W7A6
def flatten_list(lst):
    result = []
    for sub in lst:
        for item in sub:
            result.append(item)
    return result
 
a = eval(input())
print(flatten_list(a))
```

```python
# W7A7
def length_of_LIS(nums):
    n = len(nums)
    if n == 0:
        return 0

    dp = [1] * n

    for i in range(n):
        for j in range(i):
            if nums[j] < nums[i]:
                dp[i] = max(dp[i], dp[j] + 1)

    return max(dp)

a = list(map(int,input().split()))
print(length_of_LIS(a))
```

```python
#W7A8
import heapq

def min_interval(intervals, queries):
    intervals.sort()  # sort by l
    queries_with_idx = sorted((q, i) for i, q in enumerate(queries))
    
    res = [-1] * len(queries)
    heap = []  # (length, r)
    i = 0
    
    for q, idx in queries_with_idx:
        # add intervals with l <= q
        while i < len(intervals) and intervals[i][0] <= q:
            l, r = intervals[i]
            heapq.heappush(heap, (r - l + 1, r))
            i += 1
        
        # remove intervals with r < q
        while heap and heap[0][1] < q:
            heapq.heappop(heap)
        
        if heap:
            res[idx] = heap[0][0]
        else:
            res[idx] = -1
    
    return res


a = eval(input())
l = list(map(int,input().split()))
print(minInterval(a,l))
```

```python
# W7A9
from bisect import bisect_left

def countSmaller(nums):
    sorted_list = []
    result = []

    # duyệt từ phải sang trái
    for x in reversed(nums):
        # tìm số phần tử < x
        count = bisect_left(sorted_list, x)
        result.append(count)

        # chèn x vào danh sách đã sắp xếp
        sorted_list.insert(count, x)

    # đảo ngược kết quả
    return result[::-1]

a = list(map(int,input().split()))
print(countSmaller(a))
```

```python
# W7A10
def first_greater(arr, x):
    left = 0
    right = len(arr)
    while left < right:
        mid = (left + right) // 2
        if arr[mid] <= x:
            left = mid + 1
        else:
            right = mid
    return left


def maxSubarraySumModulo(nums, m):
    prefix = 0
    ans = 0
    pref = [0]   

    for x in nums:
        prefix = (prefix + x) % m

        idx = first_greater(pref, prefix)
        if idx < len(pref):
            ans = max(ans, (prefix - pref[idx] + m) % m)
        else:
            ans = max(ans, prefix)

        pref.insert(idx, prefix)

    return ans

a = list(map(int,input().split()))
x = int(input())
print(maxSubarraySumModulo(a,x))
```

```python
#W7A13
li = list(map(int, input().split()))
ans = 1
mp = {}

for i in range(len(li) - 1, -1, -1):
    cnt = max(mp.get(li[i] - 1, 0) + mp.get(li[i], 0) + 1, mp.get(li[i] + 1, 0) + mp.get(li[i], 0) + 1)
    mp[li[i]] = mp.get(li[i], 0) + 1
    ans = max(ans, cnt)

print(ans)
```
